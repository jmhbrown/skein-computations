Notebook showing the claims for different gammas, following table in other document.

```python
load("HH-of-T3.sage")
S = SL2Z.0
T = SL2Z.1
```

### For $S$ and $S^3= -S$:
A factor $2$ for each distinct $\lambda_i$, unless $\lambda_i$ is a multiple of $4$

```python
Monica_table([S,S**3],5)
```

### For $E_+ = TS$ and $E_- = (TS)^{-1}$:
A factor $1$ for each (distinct) $\lambda_i$. (Note: regular cause $|tr(id-\gamma)| = 1$)

```python
Monica_table([T*S,(T*S)**(-1)],5)
```

### For $-E_+ = -TS$ and $-E_- = -(TS)^{-1}$:
A factor $3$ for each distinct $\lambda_i$, unless $\lambda_i$ is a multiple of $3$

```python
Monica_table([S**2*T*S,S**2*(T*S)**(-1)],6)
```

### For Id:
A factor $1$ for each (distinct) $\lambda_i$.

```python
Monica_table([S**4],6)
```

### For -Id:
A factor $4$ for odd distinct $\lambda_i$, a factor $1$ for even $\lambda_i$.
Note that code gives $(2,2)$ and not $(4)$ but for calculating skein module dimensions we only care about the product of the code output.

```python
Monica_table([S**2],6)
```

### For $T^m$, $m$ nonzero integer:
A factor $|m|\lambda_i$ for each distinct $\lambda_i$.
Same remark as before: when calculating skein module dimensions we only care about the product of the code output.

```python
Monica_table([T**k for k in range(-6,0)]+[T**k for k in range(1,7)],5)
```

### For $-T^m$, $m$ nonzero integer:
A factor $4$ for each distinct odd $\lambda_i$, a factor $2$ for each distinct even $\lambda_i$.
Same remark as before: when calculating skein module dimensions we only care about the product of the code output. For example for odd $m$ it gives $(4)$ for odd $\lambda_i$ and for even $m$ it gives $(2,2)$ for odd $\lambda_i$, so they give same end result.

```python
Monica_table([S**2*T**k for k in range(-6,0)]+[S**2*T**k for k in range(1,7)],5)
```

### For regular ones, i.e. $|tr(\gamma)| > 2$ :
A factor equal to $|tr(\gamma-Id)|$ for each distinct $\lambda_i$.
Blue_table is exactly the same as Monica_table + an extra column that prints $|tr(\gamma-Id)|$. Code gives results for 10 random gammas.

```python
Blue_table([SL2Z.random_element() for k in range(10)],5)
```
